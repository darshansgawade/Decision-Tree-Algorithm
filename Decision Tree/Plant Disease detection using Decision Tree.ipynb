# Plant Disease detection for iris dataset using Decision Tree

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Loading the dataset

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

### Summarise the dataset

```python
print(data.data)
print(data.data.shape)
```

```python
print(data.target)
print(data.target.shape)
```

### Dividing the dataset into X,Y

```python
X=pd.DataFrame(data.data,columns=data.feature_names)
X
```

```python
Y=data.target
Y
```

### Spliting the dataset using train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

### Building the model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
accuracy=[]
for i in range(1,10):
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    accu=accuracy_score(y_test,pred)
    accuracy.append(accu)

plt.figure(figsize=(12,6))
plt.plot(range(1,10),accuracy,color="red",marker="o",markerfacecolor="blue",linestyle="dashed",markersize=10)
plt.title("Finding best max_depth value")
plt.xlabel("Max_depth")
plt.ylabel("Accuracy")
    
```

### Building Original Model

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy",max_depth=3,random_state=0)
```

```python
model.fit(x_train,y_train)
```

### Predicting the all values

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
accu=accuracy_score(y_pred,y_test)*100
print("Accuracy of Model is : ",accu,"%")
```
